## 2 Coin flip Analysis
### 2.1 Analysis Background
Theoretically, a coin flip's ratio of heads to tails is 1:1, or 1 to 1. This indicates that if a coin is repeatedly flipped, we should anticipate seeing roughly the same number of heads as tails. In other words, there is an approximately 50% chance that the coin lands on either side with each throw. Due to factors like the coin's bias, the flipper's ability, and other random variables, the ratio of heads to tails may fluctuate slightly in practice. However, over many flips, the heads-to-tails ratio is still close to 1:1. For example, the coin is flipped 100 times, we might see 56 heads and 44 tails. While this is not precisely a 1:1 ratio, it indicates that the coin is fair and the flips are random.

### 2.2 Cleaning and Extracting Data from 40,000 coin flips
Below is a dataset that flips a coin 40,000 times. Flipped 20,000 times Head faced up, and 20,000 times Tail faced up, respectively. The dataset contains many null boxes and unnecessary data, which require cleaning before formal analysis.

The cleaning process involved several steps: removing unwanted columns, dropping rows with empty cells, and removing decimal values from specific columns. The resulting dataset is much more organized and concise, making it easier to analyze and interpret. For example, we can use the dataset to explore patterns in the outcomes of the coin flips and see if certain tossers are more likely to get heads or tails. Overall, the cleaned dataset produces valuable information as a foundation for understanding probability.

```python
# Reads CSV file and store the data in a DataFrame
df = pd.read_csv("coinflip.csv")

df 
```

```python
list = []

# Reformat / Delete unwanted columns
df.drop_duplicates(subset=["Toss #", "Toss #.1"], inplace=True) #Removed 3 Duplicates
df = df.drop(columns=["Date", "Time", "Tosser", "Toss #", "Unnamed: 7", "Unnamed: 8","Time.1", "Tosser.1", "Toss #.1"])

# Splitting the dataset as the information presented is not usable
df_1 = df.drop(columns=["Start.1", "End: H.1", "End: T.1"])
df_2 = df.drop(columns=["Start", "End: H", "End: T"])
df_2.columns = ["Start", "End: H", "End: T"]
```

```python
# remove rows with empty cells in the "Toss #" and "Start" columns
df_1.dropna(subset=["Start"], inplace=True)
df_2.dropna(subset=["Start"], inplace=True)
df_1.fillna(0, inplace=True)
df_2.fillna(0, inplace=True)
```

```python
# Merge the data
list = pd.concat([df_1, df_2], ignore_index=True)

fin_df = pd.DataFrame(list)
fin_df['End: H'] = pd.to_numeric(fin_df['End: H'], errors='coerce')
fin_df['End: T'] = pd.to_numeric(fin_df['End: T'], errors='coerce')

# Checking if all information fits the criteria
fin_df = fin_df.loc[fin_df['Start'].str.len() <= 1]
rows_to_drop = fin_df[fin_df['End: H'] == fin_df['End: T']].index
fin_df.drop(rows_to_drop, inplace=True)

# Shows Final Processed Dataset
# fin_df [D1]
```

### 2.3 Data Examination
The cleaned data below examines the probability of a coin landing heads up or tails up. This helps to ensure the data is accurate and consistent, eliminates bias, and allows for more efficient data management and analysis.

#### Probabilities of both possible starting points
The Pie charts below show the probability of a coin landing heads up or tails up with 20,000 throws starting with "Heads" facing up and 20,000 throws starting with "Tails" facing up:

```python
df_H = fin_df[fin_df["Start"].str.contains("H")]
df_T = fin_df[fin_df["Start"].str.contains("T")]
labels = ["Ended Heads", "Ended Tails"]

# Data to plot
startH_H = df_H['End: H'].sum()
startH_T = df_H['End: T'].sum()
startT_H = df_T['End: H'].sum()
startT_T = df_T['End: T'].sum()
n = [startH_H, startH_T]
o = [startT_H, startT_T]
```

```python
#Pie Chart to show Percentage probability (Starting Heads)
plt.pie(n, labels=labels, autopct='%1.3f%%', colors=["#D2E1F1","#C492C1"])
plt.title("Flip Started with 'Heads' facing up")
plt.axis('equal')
plt.show()

#Pie Chart to show Percentage probability (Starting Tails)
plt.pie(o, labels=labels, autopct='%1.3f%%', colors=["#A2E1F1","#D492C1"])
plt.title("Flip Started with 'Tails' facing up")
plt.axis('equal')
plt.show()
```

## 3 Gambling

Gambling is the act of something of value(i.e. money) on an event or activity with an uncertain outcome. The primary intent of gambling is typically to win additional money or material goods. Gambling comes in many forms: casino games, lotteries, sports betting, and other games of chance. ### 3.2 What makes gambling profitable?
For the hosts of a gambling venue, such as a casino or online betting site, profitability is achieved through the widespread appeal of the gambling activities offered. With this, the venue can attract many players, which allows them to generate revenue through the charge for their services, such as the fees and commissions, and the profits they make from the gambling activities themselves.

The widespread availability of gaming alternatives allows gamblers to profit from their activity. This enables the person to partake in a range of gambling activities and maybe raise their winning odds. However, it is crucial to remember that gambling is customarily regarded as entertainment and most of those who bet do it for enjoyment rather than to make money. As a result, people must gamble sensibly and within their means, as it can be a risky pastime that, if not done carefully, can result in substantial financial losses.

With the odds skewed towards the hosts, there are several ways in which the gambling venue, such as a casino or online betting site, can achieve profitability over the players:
1. House edge: In many gambling activities, the operator has an inherent advantage over the player, known as the "house edge." Over time, the operator will likely profit even if some players win in the short term.

2. Fees and commissions: Many gambling venues charge fees or take a commission on the bets placed by their customers. For example, a casino may charge a fee for using certain table games or slot machines or take a percentage of players' money won or lost.

3. Non-gambling revenue: Many gambling venues, especially casinos, generate revenue from non-gambling sources, such as food and beverage sales, hotel rooms, and entertainment.

4. Volume: By attracting many players, a gambling venue can generate significant revenue through the sheer volume of gambling activity.

The profitability of a gambling venue is often dependent on a combination of these factors and external factors such as competition and economic conditions.

### 3.3 Gambling Dataset Analysis (Bustabit Gambling)
For a gambling analysis, this project will be using the Bustabit platform. Bustabit is a type of online gambling game involving betting on the outcome of a virtual "crash" of a constantly increasing number. Players place bets on the website, and the game's server generates a random number between 0 and 100. The game continues until the number "crashes," which means it falls below a certain threshold. After crashing, the game will be considered over, and all remaining bets are settled based on the multiplier at which the game crashed.

The multiplier value calculates a player's win at the cash-out timing. For example, if they bet 200 and the value was 2.50x at the time they cashed out, they win 500. In a winning game, a percentage Bonus per game is multiplied and added according to the player's bet and summed to give a final Profit. Assuming a Bonus of 1%, the Profit for this round would be (bet x 2.5) + (bet x .01) - bet, winning you 320.

The multiplier is determined by the point at which the game crashes, with higher multipliers corresponding to lower crash points. For example, if the game crashes at a multiplier of 50, all bets placed at or below that multiplier will win, while bets placed above it will lose.

```python
# Reads CSV file and store the data in a DataFrame
dfb = pd.read_csv("bustabit.csv")
dfb.fillna(0, inplace=True)
dfb = dfb.sort_values("Profit", ascending = False)
#dfb [D2]
```

#### Finding the winnings of the House vs the Players
The findings below might be misleading at first glance, showing that the winning-to-losing ratio is higher, meaning that players generally win more due to the nature of the gambling game. Players can hop off as early as they want to prevent any actual losses and take small wins at a time. However, these small winning tactics do little to the overall pool of money won by the house over the sheer amount of bets placed. The house has won much more in total bits than win/loss ratios.

```python
WinnerRows = dfb[dfb['Profit'] > 0.0]
LoserRows = dfb[dfb['Profit'] == 0.0]
tWin = WinnerRows.shape[0]
tLoss = LoserRows.shape[0]
```

```python
# Set the values and labels for the pie chart
y = [tWin, tLoss]
labels = ['Win', 'Loss']

# Create a list of strings to use in the legend
legend_labels = [f'Total Wins: {tWin}', f'Total Losses: {tLoss}']

# Plot the pie chart
plt.pie(y, labels=labels, autopct='%1.3f%%', colors=["#80ED99","#F6CA83"])
plt.legend(legend_labels, bbox_to_anchor=(1,0), loc="lower right", bbox_transform=plt.gcf().transFigure)
plt.title("Player win/loss ratio")
plt.axis('equal')
plt.show()
```

```python
# Data to plot
tProfit = dfb['Profit'].sum()
tBet = dfb['Bet'].sum()
houseProfits = tBet - tProfit
y = [tProfit, houseProfits, tBet]
labels = ["Player Winnings", "House Profits", "House total Pool"]
```

```python
# Create Bar Chart
plt.figure(figsize = (6, 6))
plt.bar(labels, y, color=["#80ED99", "#EF2E45", "#F6CA83"])
plt.title("House Winnings vs Player Winnings")
plt.xlabel("Profits")
plt.ylabel("Total Bits (Billions)")

# function to add value labels
for i, v in enumerate(y):
    label = format(v, '.2f')  # round value to 2 decimal places
    plt.text(i, v, label, color='black', horizontalalignment="center")

plt.show()

# percentage the house earned
perEarned = houseProfits / tBet
formatted_value = format(perEarned, '.5f')  # round value to 5 decimal places
print("Total Percentage that goes to the house: ", formatted_value, "%")
```

### 3.4 Data Analysis & Conclusion
The information gathered shows the drastic difference relative to total player winnings vs the house[5] and reflects similarly in many articles about gambling and how the house earns its income. The house is not a charity that gives away money; it also has its business model and is where profit models such as "house edge" shows great potential, as seen here by the 70% total percentage profits that end up going to the house.

Stacking odds of winning a profit gets smaller and smaller as time goes on as a player would have to keep winning and earning more than they will lose. The house edge is a small percentage of the bets the house expects to win in the long run. It is the difference between the true odds of an event occurring and the odds that the casino pays out.

A quick example is in a roulette game, where the odds of landing on a specific number are 37 to 1 (since there are 37 numbers on the roulette wheel). However, the casino pays out only 35 to 1 for a winning bet on a specific number, meaning that the house edge is 2/37, or approximately 5.4%.

The house edge changes depending on the game and the specific rules in place, and some games have a higher house edge than others, meaning the casino has a higher expected return on those games. In other words, the house edge represents the built-in advantage that the casino has over the players. It is how the casino makes a profit in the long run.

```python
URL = "https://www.macrotrends.net/stocks/charts/ATVI/activision-blizzard/gross-profit"
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')

# check if website is accessible
#print((soup.prettify()))
```

```python
# Find the table element
table = soup.find('table', class_='historical_data_table table')

# Initialize empty lists to store the data
dates = []
gross_profits = []

# Iterate through the rows in the table
for row in table.find_all('tr'):
  # Find the first td element in the row (the date) and extract the text
  date_td = row.find('td')
  if date_td:
    date = date_td.text
    dates.append(date)
  
  # Find the second td element in the row (the gross profit) and extract the text
  gross_profit_tds = row.find_all('td')
  if len(gross_profit_tds) > 1:
    gross_profit = gross_profit_tds[1].text
    gross_profits.append(gross_profit)

gross_profits = [profit.replace('$', '') for profit in gross_profits]

# Print the lists to verify the data was extracted correctly
# print(dates)
# print(gross_profits)
```

```python
# Convert the x and y lists to numerical data types
dates = [int(year) for year in dates]
gross_profits = [int(profit.replace(',', '')) for profit in gross_profits]
```

```python
# Plot the data
plt.plot(dates, gross_profits, color="#003DF5")

# Add dots for all points in the line graph
plt.scatter(dates, gross_profits, s = 50, c = "red")

# Add labels and title
plt.title('Gross Yearly Profit of Activision Blizzard')
plt.xlabel('Year')
plt.ylabel('Gross Profit (in millions)')

# Show the plot
plt.show()
```

```python
# Find all table elements with the specified class
all_tables = soup.find_all('table', class_='historical_data_table table')

# Access the second table element in the list
target_table = all_tables[1]

# Initialize empty lists to store the data
periods = []
profits = []

# Iterate through the rows in the table
for row in target_table.find_all('tr'):
  # Find the first td element in the row (the period) and extract the text
  period_td = row.find('td')
  if period_td:
    period = period_td.text
    periods.append(period)
  
  # Find the second td element in the row (the profit) and extract the text
  profit_tds = row.find_all('td')
  if len(profit_tds) > 1:
    profit = profit_tds[1].text
    profits.append(profit)

profits = [profit.replace('$', '') for profit in profits]

# Print the lists to verify the data was extracted correctly
# print(periods)
# print(profits)
```

```python
# Convert the profits data to integers
profits = [int(profit.replace(',', '')) for profit in profits]

# Convert the periods data to datetime objects
from datetime import datetime
periods = [datetime.strptime(period, '%Y-%m-%d') for period in periods]
```

```python
# Set the x-axis values to a sequence of evenly spaced numbers
x = range(len(periods))[::-1]

# Set the y-axis values to the profits
y = profits[::1]

# Create a figure with a width of 12 inches
plt.figure(figsize=(12, 6))

# Plot the bar chart
plt.bar(x, y, width=0.8, color='#0A47FF', label='Profits', align='edge', edgecolor='grey')
plt.legend()

# Set the x-axis tick labels to the periods
plt.xticks(x, [period.strftime('%Y-%m-%d') for period in periods], rotation=85)

plt.title("Activision Blizzard Quarterly Gross Profit")
plt.xlabel('Periods')
plt.ylabel('(Millions of US $)')
plt.show()
```

However, critics have pointed out that the random nature of loot boxes can lead to an uneven distribution of items, with some players getting lucky and receiving valuable items. In contrast, others may end up with less valuable items or duplicates. This means that a person could quickly get a high-value item at any time if they are lucky, but someone that has spent thousands of dollars would have a different luck.

Some people argue that loot boxes are a form of gambling because they involve the purchase of a chance to obtain a reward, the contents of which are unknown until opened. This element of chance can be addictive, and some people may spend large amounts of money on loot boxes to obtain rare or valuable items. Every time a player makes a wish, they have a certain percentage chance of obtaining a 5-star character or weapon. If a player does not receive a 5-star character or weapon after making a certain number of wishes, the game will increase their chances of receiving one on the next wish.
